```python
import requests
import pandas as pd
```

```python
def main():
    url = "https://api.worldbank.org/v2/country/all/indicator/FP.CPI.TOTL"

    params = {"format": "json", "date": "2010:2024", "per_page": 10000}

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
    else:
        raise RuntimeError(f"failed to fetch data: {response.status_code}")

    df = pd.DataFrame(data[1])
    df = df[pd.notna(df["value"])]
    df["contry_name"] = df["country"].apply(lambda x: x["value"])
    df["contry_id"] = df["country"].apply(lambda x: x["id"])
    df = df[["countryiso3code", "contry_name", "date", "value"]]
```
